Para el enunciado: "Manipule estos archivos para realizar lo siguiente: Generar una estructura con los nombres de los estudiantes y la suma de ambas notas."

La solución planteada a continuación comienza definiendo una lista vacía (lista_notas) donde se van a almacenar la suma de las notas, las cuales serán leídas y procesadas por una función definida como: "sumar_notas". Dicha función recibirá como parámetro dos archivos de tipo ".txt" (archivo1,archivo2)

La función itera (con el uso de una variable local "i") uno de los archivo .txt linea a linea y agrega con la función ".append()" los valores (transformados a tipo int) a una variable tipo lista (inicializada localmente a la función, la cual luego será retornada por la misma). Para ello se abre el archivo .txt desde la función (usando ".open()") en modo "read" y se almacena en la variable local "archivo_notas_1".

Nota: se hace uso de una funcion "limpiar_cadena" a la cual se le pasa como parámetro una cadena de texto y devuelve dicha cadena de forma más prolija para la salida de la terminal quitando determinados caracteres.

a continuacion se abre en modo "read" y se itera el segundo archivo .txt linea a linea y se agrega el valor (transformado a entero) a la "lista" local, sumandose con el valor que corresponde a la posición actual de la iteración.

finalmente se imprime "lista_notas" que es retornada por la función "sumar_notas" para comprobar que la suma se haya realizado correctamente.

```python
def limpiar_cadena(cadena):
    """ Devuelve una cadena de texto de forma mas legible """
    return cadena.replace(' ', '').replace('\'', '').replace(',', '').replace('\n', '').lower()

def sumar_notas(archivo1,archivo2):
    """ Suma los valores linea a linea de 2 archivos de texto que se pasan como parámetro
        y los devuelve en una lista

        returns: list
    """
    lista = []
    i = 0
    with open(archivo1, "r") as archivo_notas_1:
        for linea in archivo_notas_1:
            lista.append(int(limpiar_cadena(linea)))

    with open(archivo2, "r") as archivo_notas_2:
        for linea in archivo_notas_2:
            lista[i] = lista[i] + int(limpiar_cadena(linea))
            i += 1

    return lista

lista_notas = []
lista_notas = sumar_notas("eval1.txt","eval2.txt")

for elem in lista_notas:
    print(elem)
```

continuando con el enunciado: "Manipule estos archivos para realizar lo siguiente: Generar una estructura con los nombres de los estudiantes y la suma de ambas notas."

Se declara una funcion "notas_alumnos" la cual generará la estructura principal que recibe como parámetros: una lista que contenga la suma de todas las notas ("lista_notas") la cual fué generada anteriormente a través del uso de la función "sumar_notas" y un archivo tipo ".txt" que contenga los nombres de los alumnos ("archivo_nombres"). 

la función declara localmente una lista llamada "lista_nombres_notas" la cual será retornada al finalizar. 
Se abre el archivo .txt que se pasó como parámetro a través de la función "open()" en modo "read" y se itera almacenando en la lista "lista_nombres_notas" los nombres de los alumnos como valor de un diccionario en la clave "nombres". De igual forma se almacenan como valor del diccionario en (cada posición de una nueva iteración) en la clave "notas" los datos que se obtienen de la lista pasada por parámetro "lista_notas"

la variable "notas_todos" de tipo lista tomará el valor retornado por la función "notas_alumnos"

```python
def notas_alumnos(lista_notas,archivo_nombres):
    """Ésta función devuelve una estructura tipo lista donde cada uno de sus índices son diccionarios donde se 
       almacenan en una clave 'nombre', el nombre del alumno como valor, y en otra clave 'nota', la suma de 
       las notas de las evaluaciones 1 y 2 como valor.

       :param lista_notas: desde éste parámetro tomo los valores para agregar en las claves 'notas'
       :type lista_notas: list

       :param archivo_nombres: desde éste parametro se toman los nombres para almacenar en las claves 'nombres'
       :type archivo_nombres: txt file

       returns: list
    """
    lista_nombres_notas = []
    i = 0

    with open(archivo_nombres, "r", encoding="utf8") as nombres:
        for linea in nombres:
            lista_nombres_notas.append({"nombre": limpiar_cadena(linea)})
        for linea in lista_notas:
            lista_nombres_notas[i]["nota"] = linea
            i += 1

    return lista_nombres_notas 

notas_todos = []
lista_notas = sumar_notas("eval1.txt","eval2.txt")
notas_todos = notas_alumnos(lista_notas,"nombres_1.txt")

for elem in notas_todos:
    print(elem)
```

para la segunda parte del enunciado: "Calcular el promedio de las notas totales..."

Se definió una función "calcular_promedio" a la cual se le pasa como parámetro una lista. Dicha lista podría haber sido la lista de notas generada anteriormente (lista_notas). Pero decidí pasarle como parámentro la estructura principal (notas_todos) para probar el acceso a la clave "nota" del diccionario.  

```python
def calcular_promedio(lista):
    """Ésta función devuelve el promedio de notas desde una lista como parámetro con todas las notas """
    prom = 0
    for i in lista:
        prom = prom + i["nota"]
    prom = prom / len(lista)

    return prom

promedio = calcular_promedio(notas_todos)
print(f"el promedio de notas es: {promedio}")
```
